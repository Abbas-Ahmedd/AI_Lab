# Task 3

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4 * np.pi, 20)
y1 = np.cos(x)
y2 = (2 * x) + 1

# plt.figure(figsize=(14,7))
plt.plot(x, y1, marker="o", label="cos(x)")
plt.plot(x, y2, marker="x", label="y = 2x + 1")

plt.legend()
plt.title("Straight line and cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
