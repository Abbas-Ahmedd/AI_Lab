```python
import math
import heapq

graph = {
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Zerind': {'Arad': 75, 'Oradea': 71},
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Sibiu': {'Arad': 140, 'Oradea': 151, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Giurgiu': 90, 'Urziceni': 85}
}

lat_long = {
    'Arad': (46.1667, 21.3167), 'Bucharest': (44.4167, 26.1000),
    'Craiova': (44.3333, 23.8167), 'Drobeta': (44.6259, 22.6566),
    'Fagaras': (45.8416, 24.9730), 'Giurgiu': (43.9037, 25.9699),
    'Hirsova': (44.6833, 27.9500), 'Iasi': (47.1585, 27.6014),
    'Lugoj': (45.6904, 21.9033), 'Neamt': (46.9283, 26.3705),
    'Oradea': (47.0553, 21.9214), 'Pitesti': (44.8565, 24.8697),
    'Rimnicu Vilcea': (45.1042, 24.3758), 'Sibiu': (45.7977, 24.1521),
    'Timisoara': (45.7489, 21.2087), 'Urziceni': (44.7167, 26.6333),
    'Vaslui': (46.6333, 27.7333), 'Zerind': (46.6225, 21.5174)
}

def heuristic(city1, city2):
    lat1, lon1 = lat_long[city1]
    lat2, lon2 = lat_long[city2]
    return math.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2) * 111


class Node:
    def __init__(self, city, parent=None, g=0, h=0):
        self.city = city
        self.parent = parent
        self.g = g
        self.h = h
        self.f = g + h

    def __lt__(self, other):
        return self.f < other.f


def astar(start, goal):
    open_list = []
    heapq.heappush(open_list, Node(start, None, 0, heuristic(start, goal)))
    visited = set()

    while open_list:
        current = heapq.heappop(open_list)

        if current.city == goal:
            path = []
            distance = current.g
            while current:
                path.append(current.city)
                current = current.parent
            return path[::-1], distance

        visited.add(current.city)

        for neighbor, cost in graph.get(current.city, {}).items():
            if neighbor in visited:
                continue

            g = current.g + cost
            h = heuristic(neighbor, goal)

            heapq.heappush(open_list, Node(neighbor, current, g, h))

    return None, None


path, distance = astar("Arad", "Bucharest")

print("Shortest path from Arad to Bucharest:", path)
print("Distance:", distance)
```
