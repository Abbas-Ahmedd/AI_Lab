
# Heuristic values (straight-line distance to Bucharest)

```python

heuristic = {
    "Arad": 366,
    "Zerind": 374,
    "Oradea": 380,
    "Sibiu": 253,
    "Timisoara": 329,
    "Lugoj": 244,
    "Mehadia": 241,
    "Dobreta": 242,
    "Craiova": 160,
    "Rimnicu Vilcea": 193,
    "Fagaras": 176,
    "Pitesti": 100,
    "Bucharest": 0,
    "Giurgiu": 77,
    "Urziceni": 80,
    "Hirsova": 151,
    "Eforie": 161,
    "Vaslui": 199,
    "Iasi": 226,
    "Neamt": 234
}
```

# Best First Search

```python

def best_first_search(start, goal):
    pq = PriorityQueue()
    pq.put((heuristic[start], start))

    visited = set()
    parent = {}
    cost = {start: 0}

    while not pq.empty():
        h, current = pq.get()

        if current == goal:
            break

        visited.add(current)

        for neighbor, distance in graph[current].items():
            if neighbor not in visited:
                parent[neighbor] = current
                cost[neighbor] = cost[current] + distance
                pq.put((heuristic[neighbor], neighbor))

    path = []
    node = goal

    while node != start:
        path.append(node)
        node = parent[node]

    path.append(start)
    path.reverse()

    return path, cost[goal]


# User input
start_city = input("Enter starting city: ")
goal_city = "Bucharest"

path, distance = best_first_search(start_city, goal_city)

print("\nPath to Bucharest:")
print(" -> ".join(path))
print("Total distance:", distance, "km")
```
