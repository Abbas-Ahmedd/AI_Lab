```python
def fire_alarm_system(s, t):
    smoke = s
    temp = t
    status_alarm = 0 #off
    status_sprinkle = 0 #off
    if smoke == 1 and temp < 70:
        print("Smoke detected!")
        if sound_alarm(smoke , status_alarm) == 1:
            print("Sounding the alarm!!")
        if sprinkler_system(smoke , status_sprinkle) == 1:
            print("Activate the sprinkle system.")
    elif temp > 60:
        call_fire_department(temp)
    else:
        case_3(smoke , temp)

def sound_alarm(smoke , status_alarm):
    if smoke == 1 and status_alarm == 0:
        sound_status = 1 #on
    elif smoke == 1 and status_alarm == 1:
        sound_status = 1 #on
    elif smoke == 0 and status_alarm == 1:
        print("Alarm is not working.")
        return None
    else:
        sound_status = 0 #off
    return sound_status

def sprinkler_system(smoke , status_sprinkle):
    if smoke == 1 and status_sprinkle == 0:
        sprinkle_status = 1 #on
    elif smoke == 1 and status_sprinkle == 1:
        sprinkle_status = 1 #on
    elif smoke == 0 and status_sprinkle == 1:
        print("Sprinkle is not working.")
        return None
    else:
        sprinkle_status = 0 #off
    return sprinkle_status

def call_fire_department(temp):
    smoke_1 = 1
    status_alarm = 1
    status_sprinkle = 1
    if temp > 60:
        if sound_alarm(smoke_1 , status_alarm) == 1:
            print("Sounding the alarm!!")
        if sprinkler_system(smoke_1 , status_sprinkle) == 1:
            print("Activate the sprinkle system.")
        print("Calling the fire department!")
    return

def case_3(smoke , temp):
    if smoke == 0 and temp < 70:
        status_alarm = 0 #off
        status_sprinkle = 0 #off
        print("Normal state.")

if __name__ == "__main__" :
    smoke = int(input("Did you see smoke? (0/1): "))
    temp = int(input("Enter the temperature of the office: "))
    fire_alarm_system(smoke , temp)
```

# Task 2

```python
import logging

class AutomaticWateringSystem:
    def __init__(self, dry_threshold=30, moist_threshold=70, wet_threshold=90):
        self.dry_threshold = dry_threshold
        self.moist_threshold = moist_threshold
        self.wet_threshold = wet_threshold
        self.watering_system = False

    def determine_soil_state(self, moisture_level):
        if moisture_level < self.dry_threshold:
            return "Dry soil"
        elif moisture_level >= self.dry_threshold and moisture_level < self.moist_threshold:
            return "Moist soil"
        elif moisture_level >= self.moist_threshold and moisture_level < self.wet_threshold:
            return "Wet soil"
        else:
            logging.error("Invalid moisture level")
            raise ValueError("Invalid moisture level")

    def control_watering_system(self, moisture_level):
        soil_state = self.determine_soil_state(moisture_level)
        if soil_state == "Dry soil":
            self.watering_system = True
        elif soil_state == "Moist soil":
            self.watering_system = False
        elif soil_state == "Wet soil":
            self.watering_system = False
        else:
            logging.error("Invalid soil state")
            raise ValueError("Invalid soil state")

if __name__ == "__main__":
    logging.basicConfig(filename="watering_system.log", level=logging.ERROR)
    system = AutomaticWateringSystem()
    try:
        system.control_watering_system(25)
        print("Watering system: ", system.watering_system)
        system.control_watering_system(45)
        print("Watering system: ", system.watering_system)
        system.control_watering_system(95)
        print("Watering system: ", system.watering_system)
    except ValueError as error:
        print("Error: ", error)
```
